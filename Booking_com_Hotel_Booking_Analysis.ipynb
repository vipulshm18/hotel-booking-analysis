## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df = df.drop_duplicates()
````

````python
df.loc[:,'country'] = df['country'].fillna('Unknown')
````

````python
df.loc[:,'agent'] = df['agent'].fillna(0)
````

````python
df.loc[:, 'children'] = df['children'].fillna(0)
````

````python
df.loc[:, 'children'] = df['children'].astype(int)
````

````python
df.loc[:,'company'] = df['company'].fillna(0)
````

````python
df.isna().sum()
````

````python
df['total_days'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
````

````python
df['total_guests'] = df['adults'] + df['children'] + df['babies']
````

````python
df['is_family'] = ((df['children'] + df['babies'])>0).astype(int)
````

````python
df['total_revenue'] = df['adr'] * df['total_days']
````

### What all manipulations have you done and insights you found?

Duplicate rows were removed from the dataset during the data cleaning process. Missing values were not dropped immediately, as removing observations could result in information loss and potentially introduce bias into the dataset distribution. Instead, missing values in categorical features were imputed with 'Unknown', while missing values in numerical features were filled with 0 based on the business context of the variables. Additionally, several new features were engineered to enhance the analysis, including estimated revenue, total number of guests per booking, family indicator (is_family), and total duration of stay for each booking. The dataset is only till August 2017.

#### Chart - 2

````python
# Chart - 2 visualization code
fig, ax = plt.subplots(figsize=(8,5))

sns.countplot(data=df, x='is_canceled', hue='hotel', ax=ax)

total = len(df)

for p in ax.patches:

    if p.get_height() == 0:
        continue

    percentage = f'{100 * p.get_height() / total:.1f}%'

    x = p.get_x() + p.get_width() / 2
    y = p.get_height()

    ax.text(x, y, percentage,
            ha='center', va='bottom')

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

The aim to select this chart is to see the number of canceled bookings. The number of canceled booking is more than 18% for City hotel which should be reduced to maximize the profit.

##### 2. What is/are the insight(s) found from the chart?

The number of bookings being canceled is pretty high, now this needs to be examined if a reason could be found out to ensure why these booking were canceled but the dataset doesn't include any feature that mentions the reason of these cancellation.

#### Chart - 4

````python
# Chart - 4 visualization code
plt.title('Number of bookings across different year')
ax = sns.countplot(data=df, x='arrival_date_year', hue='hotel')
plt.xlabel('Year of the bookings')
plt.ylabel('Count of the bookings')
for container in ax.containers:
  ax.bar_label(container)
````

##### 1. Why did you pick the specific chart?

The aim to select this chart was to check how many bookings were made during the different years in different hotel types.

##### 2. What is/are the insight(s) found from the chart?

The number of bookings for both hotel types was relatively similar in 2015. However, in 2016, City Hotels experienced a significant increase in bookings, with total bookings exceeding those of Resort Hotels by more than 1.5 times. A similar trend continued in 2017, where City Hotels maintained substantially higher booking volumes compared to Resort Hotels. Latest date in the dataset is August 31, 2017 the comparision between 2016 as a complete year will not be valid because 2017 has only 8 months data.

````python
fig, ax = plt.subplots(1,2, figsize=(12,4))
sns.countplot(
    data=df[(df['arrival_date'] <= '2016-08-31') & (df['arrival_date_year']==2016)],
    x='arrival_date_year',
    hue='hotel',
    ax=ax[0]

)
for container in ax[0].containers:
    ax[0].bar_label(container)
sns.countplot(
    data=df[df['arrival_date_year']==2017],
    x='arrival_date_year',
    hue='hotel',
    ax=ax[1]
)
for container in ax[1].containers:
    ax[1].bar_label(container)

plt.tight_layout()
plt.show()

````

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, the insights dervied from the booking trend can help creating positive impacts in so many ways. The significantly higher number in booking of the City hotel in both 2016 and 2017 indicates a strong demand for city-based accomodation. This insight can help the hotel management focus more on operational planning, staffing and resource allocation to maximize the occupancy and revenue. Understanding the yearly booking patterns enables hotels to predict the demand and optimize pricing strategy during the high demand periods. Aditionally, the trend in 2017 is showing a similar pattern as in 2016. The number of booking till Jan-Aug in both the year is almost same and this insight can help predict the demand for the remaining months of the 2017 which can help hotel management to focus on operational planning accordingly to maximize the occupancy hence maximizing the revenue.

#### Chart - 6

````python
daily_adr = df.groupby(['reservation_status_date', 'hotel'])['adr'].mean().reset_index()
````

````python
# Chart - 6 visualization code
plt.figure(figsize=(12,5))

sns.lineplot(
    data=daily_adr,
    x='reservation_status_date',
    y='adr',
    hue='hotel'
)

plt.title('ADR Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Average ADR')

plt.show()
````

##### 1. Why did you pick the specific chart?

The chart is chosen to gain the insights about the average daily rate across different hotel types.


##### 2. What is/are the insight(s) found from the chart?

Both hotel types shows recurring peaks and dips, indicating the room prices vary seasonally and don't remain constant throughout the year. This suggests the room pricing is influenced by the travel demand and off season reduces the ADR.

Resort hotels shows much higher volatility in ADR this could be because the Resort hotel pricing is more sensitive to the seasonal demand.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

The observed seasonal fluctuations in the ADR indicate the hotel pricing is strongly influenced by the customer demand patterns. Resort Hotels, in particular, demonstrates higher ADR peaks during certain period suggesting opportunity for dynamic pricing strategies duruing the high demand season to maximise the revenue.

The sharp fluctuations in the ADR for Resort Hotels suggests that their revenue is highly dependent on seasonal demand and during off season the ADR declines significantly which may result in negative effect on revenue generation and occupancy rates. This indicates higher vulnerability of Resort Hotels to seasonality and changing travel patterns.
